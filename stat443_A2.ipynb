```r
library(forecast)
library(tseries)

# seed for reproducibility
set.seed(123)

phi <- 0.4
theta <- 0.6

lags <- 20
theoretical_acf <- ARMAacf(ar = phi, ma = theta, lag.max = lags)
theoretical_pacf <- ARMAacf(ar = phi, ma = theta, lag.max = lags, pacf = TRUE)

plot(theoretical_acf, type = "h", main = "Theoretical ACF of ARMA(1,1)", xlab = "Lag", ylab = "ACF")
abline(h = 0)

plot(theoretical_pacf, type = "h", main = "Theoretical PACF of ARMA(1,1)", xlab = "Lag", ylab = "PACF")
abline(h = 0)



```

```r
# Load necessary library
library(forecast)

# Define parameters
phi <- 0.4
theta <- 0.6

# Calculate and plot theoretical ACF and PACF for an ARMA(1,1)
acf_values <- ARMAacf(ar = phi, ma = theta, lag.max = 20, pacf = FALSE)
pacf_values <- ARMAacf(ar = phi, ma = theta, lag.max = 20, pacf = TRUE)

# Plotting
par(mfrow = c(1, 2))  # Set up the plot area
plot(acf_values, type = "h", main = "Theoretical ACF", xlab = "Lag", ylab = "ACF")
plot(pacf_values, type = "h", main = "Theoretical PACF", xlab = "Lag", ylab = "PACF")

# Set seed for reproducibility
set.seed(123)

# Simulate ARMA(1,1) series
n <- 40
arma_process <- arima.sim(n = n, model = list(ar = phi, ma = theta))

# Plotting the simulated series
plot(arma_process, main = "Simulated ARMA(1,1) Series")

# Calculate and plot sample ACF and PACF
par(mfrow = c(1, 2))
Acf(arma_process, main = "Sample ACF")
Pacf(arma_process, main = "Sample PACF")



# Set seed for reproducibility
set.seed(123)

# Simulate ARMA(1,1) series with n = 400
n <- 400
arma_process_large <- arima.sim(n = n, model = list(ar = phi, ma = theta))

# Plotting the simulated series
plot(arma_process_large, main = "Simulated ARMA(1,1) Series with n = 400")

# Calculate and plot sample ACF and PACF
par(mfrow = c(1, 2))
Acf(arma_process_large, main = "Sample ACF with n = 400")
Pacf(arma_process_large, main = "Sample PACF with n = 400")
```

```r
library(forecast)
library(tseries)


analyze_dataset <- function(data, dataset_name) {

  plot.ts(data, main = paste("Time Series Plot -", dataset_name),
  ylab = "Value", xlab = "Time")


  acf(data, main = paste("Sample ACF -", dataset_name))
  pacf(data, main = paste("Sample PACF -", dataset_name))


  fit <- auto.arima(data,seasonal = FALSE)
  print(summary(fit))

  print('RESIDUALS')
  checkresiduals(fit)

  return(fit)
}


data1 <- read.csv("/content/Assignment2_data1.csv", header = TRUE)
data2 <- read.csv("/content/Assignment2_data2.csv", header = TRUE)
data3 <- read.csv("/content/Assignment2_data3.csv", header = TRUE)


fit1 <- analyze_dataset(data1[,1], "Dataset 1")
fit2 <- analyze_dataset(data2[,1], "Dataset 2")
fit3 <- analyze_dataset(data3[,1], "Dataset 3")
```
