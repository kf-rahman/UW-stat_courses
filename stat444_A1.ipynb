```r
# Install packages if not already installed
if (!require("glmnet")) install.packages("glmnet", dependencies = TRUE)
if (!require("genridge")) install.packages("genridge", dependencies = TRUE)

# Load packages
library(glmnet)
library(genridge)

# Load the data
data(prostate, package = "genridge")

# Prepare data
x <- as.matrix(prostate[, !(names(prostate) %in% c("lcavol", "train"))])
y <- as.matrix(prostate$lcavol)

# Fit Ridge Regression Model
ridge_model <- glmnet(x, y, alpha = 0)

# Fit LASSO Model
lasso_model <- glmnet(x, y, alpha = 1)

# Plot Ridge Regression Model
plot(ridge_model, xvar = "lambda", label = TRUE)
title("Ridge Regression Coefficients Path")

# Plot LASSO Model
plot(lasso_model, xvar = "lambda", label = TRUE)
title("LASSO Coefficients Path")
```

```r
# Install and load required packages
if (!require("glmnet")) install.packages("glmnet", dependencies = TRUE)
if (!require("genridge")) install.packages("genridge", dependencies = TRUE)
library(glmnet)
library(genridge)

# Load the prostate dataset
data(prostate, package = "genridge")

# Prepare the data
# Selecting lcavol as the response and all other variables (except 'train') as predictors
response <- prostate$lcavol
predictors <- prostate[, !(names(prostate) %in% c("lcavol", "train"))]

# Fit the ridge regression model using glmnet (alpha = 0 for ridge regression)
set.seed(123)  # For reproducibility
ridge_model <- glmnet(x = as.matrix(predictors), y = response, alpha = 0)

# Print the model summary
print(ridge_model)

# Print coefficients at default lambda
print(coef(ridge_model))

# Plot the solution path of the coefficients
plot(ridge_model, xvar = "lambda", label = TRUE)
```
